## Introduction
The **Linnerud dataset** is a valuable collection of exercise-related features and physiological measurements. In this report, we explore the dataset, analyze its structure, and discuss our findings.

## Dataset Overview
- **Purpose**: The dataset aims to understand the relationship between exercise variables and physiological outcomes.
- **Features (Data)**:
    - **Weight**: Weight during exercise.
    - **Waist**: Waist circumference during exercise.
    - **Pulse**: Pulse rate during exercise.
- **Target (Physiological Measurements)**:
    - **Weight**: Individual's weight.
    - **Waist**: Waist circumference.
    - **Pulse**: Pulse rate.

```python
from sklearn.datasets import load_linnerud
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# load dataset
linnerud = load_linnerud()
```

```python
# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(linnerud.data, linnerud.target, test_size=0.2, random_state=42)
```

```python
# create linear regression model
regr = LinearRegression()
```

```python
# train the model using training sets
regr.fit(X_train, y_train)
```

```python
# make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
# evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean squared error is {mse:.2f}")
```

```python
# extracting features (waist) and target (situps)
waist = linnerud.data[:, 1]
situps = linnerud.target[:, 0]
```

```python
from matplotlib import pyplot as plt

# creating a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(waist, situps, color= 'b', marker= 'o', label=  'Waist vs Situps')
plt.xlabel('waist circumference')
plt.ylabel('number of situps')
plt.title('linnerud dataset: waist vs situps')
plt.grid(True)
plt.legend()
plt.show()
```
